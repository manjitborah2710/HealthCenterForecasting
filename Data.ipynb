```python
import pandas as pd
import numpy as np
from pandas import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def parser(x):
    return datetime.strptime(x,'%d-%m-%Y')
```

```python
data = pd.read_csv("prepared data/DATES_FILLED_jun18-jun19.csv", index_col=0, parse_dates=[0], date_parser=parser)
data = data.dropna()
data
```

```python
data.to_csv('data.csv')
```

```python
total_freq = data.sum()
total_freq
```

```python
total_freq.max()
```

```python
data_month_freq = data.groupby([lambda x: x.year, lambda x: x.month]).sum()
data_month_freq
```

```python
total_freq.sort_values(ascending=False)
```

```python
data = data.drop(columns=['Whooping cough', 'weakness'])
corrltn = data.corr()
corrltn
```

```python
sns.set(style="white")
```

```python
mask = np.triu(np.ones_like(corrltn, dtype=np.bool))
```

```python
f, ax = plt.subplots(figsize=(20, 20))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corrltn, mask=mask, cmap=cmap, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
