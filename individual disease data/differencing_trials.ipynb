```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import datetime
```

```python
def parser(x):
    return datetime.strptime(x,'%Y-%m-%d')
```

```python
df_acidity = pd.read_csv('Acidity.csv', index_col=[0], date_parser=parser, parse_dates=[0])
df_acidity.head()
```

```python
df_acidity.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(df_acidity)
```

```python
df_acidity.shift(1).head()
```

```python
df_acidity_diff=df_acidity.diff(periods=1)
```

```python
df_acidity_diff.plot()
```

```python
plot_acf(df_acidity_diff)
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(df_acidity_diff)
```
