```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import datetime
```

```python
def parser(x):
    return datetime.strptime(x, '%Y-%m-%d')
```

```python
df_urti = pd.read_csv('Urti.csv', index_col=[0], date_parser = parser, parse_dates=[0])
df_urti.head()
```

```python
##daily plot

df_urti.plot()
```

```python
df_urti_weeklyS = df_urti['Urti'].resample('W').sum()
df_urti_weeklyM = df_urti['Urti'].resample('W').mean()
```

```python
#weekly sum plot
df_urti_weeklyS.plot()
```

```python
#weekly mean plot
df_urti_weeklyM.plot()
```

```python
#autocorrelation plot daily
pd.plotting.autocorrelation_plot(df_urti['Urti'])
```

```python
pd.plotting.autocorrelation_plot(df_urti)
```

```python
#autocorrelation plot weekly
pd.plotting.autocorrelation_plot(df_urti_weeklyM)
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df_urti)
```

```python
plot_acf(df_urti_weeklyM)
```
