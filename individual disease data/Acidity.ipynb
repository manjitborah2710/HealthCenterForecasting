```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import datetime
%matplotlib inline
```

```python
def parser(x):
    return datetime.strptime(x, '%Y-%m-%d')
```

```python
df_acidity = pd.read_csv('Acidity.csv', index_col=[0], date_parser = parser, parse_dates=[0])
df_acidity.head()
```

```python
##daily plot

df_acidity.plot()
```

```python
df_acidity_weeklyS = df_acidity['Acidity'].resample('W').sum()
df_acidity_weeklyM = df_acidity['Acidity'].resample('W').mean()
```

```python
#weekly sum plot
df_acidity_weeklyS.plot()
```

```python
#weekly mean plot
df_acidity_weeklyM.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_acf(df_acidity)
```

```python
plot_pacf(df_acidity)
```

```python
pd.plotting.autocorrelation_plot(df_acidity)
```

```python
pd.plotting.autocorrelation_plot(df_acidity_weeklyM)
```
