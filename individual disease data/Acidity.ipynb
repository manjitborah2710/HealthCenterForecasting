```python
import pandas as pd
from pandas import datetime
```

```python
def parser(x):
    return datetime.strptime(x, '%Y-%m-%d')
```

```python
df_acidity = pd.read_csv('Acidity.csv', index_col=[0], date_parser = parser, parse_dates=[0])
df_acidity.head()
```

```python
type(df_acidity)
```

```python
df_temp = df_acidity['Acidity']
type(df_temp)
```

```python
df_temp.index = pd.DatetimeIndex(df_temp.index)
df_temp.index
```

```python
idx = pd.date_range('2018-06-20', '2018-12-31', name="Date")
idx
```

```python
df_temp = df_temp.reindex(idx, fill_value=0)
```

```python
df_acidity = df_temp.to_frame()
df_acidity.to_csv('Acidity.csv')
df_acidity.head()
```

```python
##daily plot

df_acidity.plot()
```

```python
df_acidity_weeklyS = df_acidity['Acidity'].resample('W').sum()
df_acidity_weeklyM = df_acidity['Acidity'].resample('W').mean()
```

```python
#weekly sum plot
df_acidity_weeklyS.plot()
```

```python
#weekly mean plot
df_acidity_weeklyM.plot()
```
